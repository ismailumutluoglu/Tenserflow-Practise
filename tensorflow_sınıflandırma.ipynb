```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.corr()["Type"].sort_values().plot(kind='bar', figsize=(10, 6))
```

```python
y = df["Type"].values
x = df.drop(["Type"], axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_text, y_train , y_text = train_test_split(x,y,test_size = 0.3, random_state = 15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_text)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Activation , Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units = 30,activation="relu"))
model.add(Dense(units = 15,activation="relu"))
model.add(Dense(units = 15,activation="relu"))
model.add(Dense(units = 1,activation="sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer="adam")
```

```python
model.fit(x = x_train , y = y_train , epochs = 700 , validation_data=(x_test, y_text),verbose=1)
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units = 30,activation="relu"))
model.add(Dense(units = 15,activation="relu"))
model.add(Dense(units = 15,activation="relu"))
model.add(Dense(units = 1,activation="sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer="adam")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss", mode = "min" , verbose=1 , patience=25)
```

```python
model.fit(x = x_train , y = y_train , epochs=700, validation_data=(x_test, y_text), verbose=1, callbacks=[earlyStopping])
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units = 30,activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 15,activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 15,activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(units = 1,activation="sigmoid"))

model.compile(loss = "binary_crossentropy", optimizer="adam")
```

```python
model.fit(x = x_train , y = y_train , epochs=700, validation_data=(x_test, y_text), verbose=1, callbacks=[earlyStopping])
```

```python
kayipDf = pd.DataFrame(model.history.history)
```

```python
kayipDf.plot()
```

```python
pred_probs = model.predict(x_test)
tahminler = (pred_probs > 0.5).astype(int)
tahminler
```

```python
# Test verisindeki gerçek sınıf dağılımı
print('Gerçek y_text dağılımı:', np.unique(y_text, return_counts=True))

# Tahmin olasılıklarının ilk 10 tanesini göster
print('Tahmin olasılıkları (ilk 10):', pred_probs[:10].reshape(-1))
```
